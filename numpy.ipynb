```python
#real world example for vstacking

# Each row = one student
# Columns = [ID, GPA, Credits]

campus1 = np.array([
    [101, 3.2, 45],
    [102, 3.5, 60],
    [103, 2.9, 40]
])

campus2 = np.array([
    [201, 3.8, 55],
    [202, 3.1, 48],
    [203, 3.6, 43]
])
total_data = np.vstack((campus1, campus2))
total_data
```

```python
np.hsplit(total_data,3) #column spliting
```

```python
np.vsplit(total_data, 2)
```

```python
# fency indexing helps in getting specific rows or column which was not possible through slice
# as in sclicing only can get continous rows or columns i.e [3:5] etc 3 colomn to four column etc
# in fency indexing we can get specific rows and columns as follow
total_data[[2,3,5]] #to get 2,3, and 5 row
total_data[:,[1,2]] #to get 1,2 columns 


```
