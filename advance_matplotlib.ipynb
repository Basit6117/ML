# anotation 

```python
#in anotation we place a text on graph i.e if we want that every dot in scatter what that dot represent
batters = pd.read_csv('matplotlib_datasets/batter.csv')
sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
plt.figure(figsize=(18,9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'],s=sample_df['runs'])  # s=sample_df['runs'] -> this show how many runs a bater scored
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#simple example
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5, 'point 1')
plt.text(2,6, 'point 2')
plt.text(3,7, 'point 3')
plt.text(4,8, 'point 4', fontdict={'size': 12,'color':'brown'})
```

#### horizantal and vertical lines

```python
plt.figure(figsize=(18,9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'],s=sample_df['runs'])  # s=sample_df['runs'] -> this show how many runs a bater scored
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.axhline(130, color='red') #draw a horizantal red line in the 130 strike rate
plt.axvline(30, color='red') #draw a vertical red line in the avg 30 
plt.axhline(140, color='green') #exceptional hitter
```

# subplots

```python
#subplot => we can plot multiple plots combinly

# a diffrent way to plot graphs
batters
plt.figure(figsize = (15,6))
plt.scatter(batters['avg'], batters['strike_rate']) #with simple scatter way
plt.title('something')
plt.xlabel('avg')
plt.ylabel('SR')
plt.show()
```

```python
# with subplot 
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'], batters['strike_rate'],color='red',marker='D')
ax.set_title('something') # to set title
ax.set_xlabel('avg')
ax.set_ylabel('SR')
fig.show
```

```python
#why subplot technique is used => ploting  multiple graph side by side
plt.subplots(nrows=2,ncols=1)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1 ,sharex=True) # as avg on x axis is common so remove from top through sharex=True 
ax[0].scatter(batters['avg'], batters['strike_rate'])
ax[1].scatter(batters['avg'], batters['runs'])
ax[0].set_title('Avg and Strike Rate')
ax[1].set_title('Avg and Runs')
ax[1].set_xlabel('Avg')
ax[0].set_ylabel('strike rate')
ax[1].set_ylabel('Runs')
```

```python
# 4 plots
fig, ax = plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(batters['avg'], batters['strike_rate'])
ax[0,1].scatter(batters['avg'], batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
#3rd way of ploting graph
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'], batters['runs'])
ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])
```

### 3D surface plot

```python
x = np.linspace(-10,10, 100)
y = np.linspace(-10,10, 100)
xx,yy = np.meshgrid(x,y) #to create a grid search for more

z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
z = np.abs(xx) + np.abs(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
```

### contour graph

```python
#this graph is 3D below it convert to 2D plot through contour graph
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour graph => To represent 3D graph in 3D | contourf is its beutiful looking graph mean cotour filled like below in next cell
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') #contourf
fig.colorbar(p)
```

### heatmap

```python
#heatmap => depicts values for a main variable of interest across two axis variables as a grid of colored squares.
```

```python
ipl_del = pd.read_csv('matplotlib_datasets/IPL_Ball_by_Ball_2008_2022.csv')
ipl_del.head(100)
# gr = plt.plot(top['overs'], top['ballnumber'])
every_six = ipl_del[(ipl_del['ballnumber'].isin([1,2,3,4,5,6])) & (ipl_del['batsman_run'] == 6)]
every_six
```

```python
grid = every_six.pivot_table(index='overs', columns='ballnumber',values='batsman_run', aggfunc='count')
grid
```

```python
#heatmap graph
plt.figure(figsize=(15,7))
plt.yticks(ipl_del['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.imshow(grid)
plt.colorbar()
```
