# filtering FataFrame

```python
ipl.sample(10)
```

```python
#find all the final winners of ipl
# identify in the MatchNumber column that which row have the value of 'Final'
mask = ipl['MatchNumber'] == 'Final' # true for all final matches
new_df = ipl[mask] # it will fetch all the final matches 
new_df[['Season', 'WinningTeam']] #show season and winning

#if write in one line
ipl[ipl['MatchNumber'] == 'Final'][['Season', 'WinningTeam']]
```

```python
# how many super over finishes have occured
super_over = ipl[ipl['SuperOver']  == 'Y'].shape #14 rows mean 14 super over matches occured or we can do like shape[0] so it will give only rows count
super_over
```

```python
#how many matches has csk won in Kolkata
ipl[(ipl['WinningTeam'] == 'Chennai Super Kings') & (ipl['City'] == 'Kolkata')].shape[0] #so csk has won five matches in Kolkata
```

```python
#toss winner is match winner in percentage
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0] / ipl.shape[0] * 100
```

```python
# movies rate with higher than 8 and number of votes is 10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)]
```

```python
#action movies with rating higher than 7.5
movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating'] > 7.5)]
```

# adding new columns

```python
# add new column
movies['Country'] = 'india'
movies.head(2)
```

```python
#from existing ones

movies.dropna(inplace=True)
movies['lead actor'] = movies['actors'].str.split('|').apply(lambda x: x[0]) # it will givwes an error because there are missing values and pandas treat as float
                                                       # with missing values and to remove missing values we use dropna
movies.sample(2)
```

```python
#astype
ipl.head(2)
ipl.info()
ipl['ID'] = ipl['ID'].astype('int32')
ipl.info()
```

```python
#astype
ipl['Season'] = ipl['Season'].astype('category') #cahnge datatype to category to save memory
ipl.info()
```

```python
#value_counts
ipl['TossDecision'].value_counts().plot(kind='pie')

```

```python
#how many matches each team has played
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending=False) #Total matcehs played by each team
```
