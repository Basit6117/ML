```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('matplotlib_datasets/iris.csv')
iris.sample(7)
```

```python
plt.style.available
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('sepal length')
plt.ylabel('petal length')
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'],cmap='summer',alpha=0.8) # cmap='jet',cmap='winter' etc these are colors theme
plt.xlabel('sepal length')                                                               # alpha=0.8 ->this is for opacity
plt.ylabel('petal length')
plt.colorbar()
```

#### plot size

```python
plt.figure(figsize=(15,7)) #to increase length and width of plot

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'],cmap='summer',alpha=0.8) # cmap='jet',cmap='winter' etc these are colors theme
plt.xlabel('sepal length')                                                              
plt.ylabel('petal length')
plt.colorbar()
```
