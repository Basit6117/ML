```python
import pandas as pd
import numpy as np
```

```python
subs = pd.read_csv('subs.csv')
subs_ser = subs.squeeze()
```

```python
#drop_duplicates() # it drop all duplicates values  by default it delete first ocuurence we cahnge it to last occurance
temp = pd.Series([1,1,1,2,3,5,6,7,7,9,0,0])
temp.drop_duplicates()
```

```python
temp.drop_duplicates(keep="last") #it will keep last and remove first occurance
```

```python
movies_ser.drop_duplicates() #it will show only one movie fo each actors
```

```python
temp.duplicated().sum() #to determine that how many duplicates are in this case there is 4 duplicates
```

```python
#if deal with missing values
temp = pd.Series([1,2,np.nan,4, np.nan,np.nan, 7, 8, np.nan, 10])
temp
```

```python
temp.size #count missing values also (10) which is total
temp.count() #missing values not included  (6)
```

```python
#isnull -> to count missing values
temp.isnull().sum() # 4 null values
```

```python
#dropna -> it will remove all missing values
temp.dropna() 
```

```python

#fillna -> it will not replace. here we can replace with values
temp.fillna(0) #it will fill all missing values with zeros
temp.fillna(temp.mean()) #it will fill with mean
```

```python
#isin -> through it we can check multiple multiple things
vk[(vk == 49) | (vk == 100)] #painful if need to check multiple conditions 
```

```python
vk[vk.isin([49,50,100])] #easliy done through isin
```

```python
#apply ->we can apply custom logics
#if i need first name of actors only and must be in uppercase
movies_ser.apply(lambda x: x.split()[0].upper())
```

```python
#let's suppose in subs series i want to know that if i have got more subscriber than average so 
 #that will be good day other wise it will be bad day
subs_ser.mean() #average is 135

subs_ser.apply(lambda x: 'Good day ' if x > subs_ser.mean() else 'Bad day')
```
