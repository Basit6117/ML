```python
# A Pandas Series is like a column in a table. It is a 1-D array holding  a data of any type.
# 1-D mean it is like a one column of table i.e name column
```

```python
import numpy as np
import pandas as pd
```

# Series from list

```python
# string
country = ["Pakistan","Turkey", "India", "KSA", "Afg","Australia"]
pd.Series(country)
```

```python
#integer
runs = [30,40,90,58]
pd.Series(runs)
```

```python
#custom index
marks = [50,78,69,80,74]
subjects = ["English","Maths","Machine Learning", "Software Eng", "Arabic"]
pd.Series(marks, index=subjects)
```

```python
#setting a name to series object
marks = pd.Series(marks, index=subjects, name="Husnain ki marks")
marks
```

# Series from dictionary

```python
marks = {
    "Maths": 78,
    "Urdu": 90,
    "Computer Science": 87,
    "Pak Study": 46,
}
marks_series = pd.Series(marks, name="Saad ki marks")
marks_series
```

# Series using read_csv

```python
#with one column
pd.read_csv('subs.csv')  #read_csv always store data in a dataframe not in series
type(pd.read_csv('subs.csv')) #DataFrame
```

```python
# to convert to series we were  added squeeze=True now we are using squeeze separately like below
import pandas as pd
subs = pd.read_csv('subs.csv')
subscriber = subs.squeeze("columns")
type(subscriber)
```

```python
# with two column 
kohli_scores = pd.read_csv('kohli_ipl.csv',index_col="match_no")
scores_Series = kohli_scores.squeeze() #convert to Series
scores_Series
```

```python
movies = pd.read_csv("bollywood.csv",index_col='movie')
boll_movies = movies.squeeze() 
print(boll_movies)
boll_movies.name #name will always be the values's column name in this case lead is the second column name
```

# Series methods

```python
# head and tail

boll_movies.head() # head by default give first columns but we can also specify how many columns we needed like in next cell
```

```python
boll_movies.head(3) # 1st three mvoies
```

```python
#tail gives last 5 columns
boll_movies.tail() 
boll_movies.tail(10) #similarly last 10 movies 
```

```python
#sample
    #gives randomly one row
boll_movies.sample() 
boll_movies.sample(4) #gives randomly 4 rows 
```

```python
#value_count => frequency count of every value
    #example: If i need to know that how many movies shah rukh khan have done

boll_movies.value_counts() #gives all actor's movies in decending order
```

```python
#sort_values -> inplace
    #sort_values sort all the values for example we need highest score of Kohli in IPL so we can sort it etc
# it does not change the original series
scores_Series.sort_values()  #gives in decending order 
#scores_Series.sort_values(ascending = False)  # Now from highest to lowest 
#scores_Series.sort_values(ascending = False).head(1) #provide highest score
#scores_Series.sort_values(ascending = False).head(1).values #provide only values which is score's column
#scores_Series.sort_values(ascending=False).head(1).values[0] # this whole process doing is called method chaining
```

```python
 # scores_Series.sort_values(inplace = True) # changes in permanent series
```

```python
# sort_indexe => sort the index
boll_movies.sort_index() #sort the index in this case movies names are indexes sorted in ascending order
```

# Series Maths Method

```python
# count -> count gives total no of items present in the series
# for example: to know that how many matches played by virat kohli
scores_Series
scores_Series.count() # 215 matches played by virat kohli
# scores_Series.size # 215

# Note: size also work the same job like count but size count missing values also. Count does not count missing values
```

```python
# sum, product
subs.sum() # it will sum all the subscriber in the subs series have
# subs.product()
```

```python
# mean-> median -> mode -> std -> var

#subscribers gain every day
subs.mean()
scores_Series.median() #median scores by virat kohli is 24

boll_movies.mode() #mode gives the value that is more repeated in this case Akshy Kumar mean he done most of the movies 

# scores_Series.std() #statistical method
scores_Series.var() #statistical method
```

```python
# min max
subs.max() #maximum subscriber gain in a single day
subs.min() # minimum subscraiber gain in a single day
```

```python
# describe  -> gives summary of the whole series i.e count,mean, std,min,max of the series
scores_Series.describe()

```

# Series indexing

```python
#integer indexing
x = pd.Series([2,3,4,7,98,56,3])
x[4] #indexing is same as in python but here negetive indexing is not valid
```

```python
# negetive indexing
# x[-3] #negetive indexing will through an error in  series read more about position and label (position mean 0,1,2,3 the indexing by default and  label mean the value that we are providing as index in movie case it is movie title )
```

```python
boll_movies.iloc[-1]  # use this rather than only simple indexing so it will not through any error
```

```python
# slicing
scores_Series[5:16] #virat kohli scores from 5 to 16 match
```

```python
boll_movies[-5:] # bollywood last 5 movies
```

```python
# fency indexing
scores_Series[[2,4,6,8]] # kohli's 2,4,6,8 matches scores
```

```python
#indexing with labels ->fency indexing
# both are valid
boll_movies['Awara Paagal Deewana']

boll_movies.loc['Awara Paagal Deewana']
```

```python
# find no of 50's and greater
scores_Series[scores_Series>=50].size
```

```python
#Find no of ducks
scores_Series[scores_Series == 0].size
```
