```python
#from a pandas dataframe
batsman = pd.read_csv('matplotlib_datasets/sharma-kohli.csv')
batsman
```

```python
#from a pandas datafram
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
#ploting multiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
#labels title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# color(hex) 
plt.plot(batsman['index'], batsman['V Kohli'],color='#03fcf0')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fc03db')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#lineStyle => dashed, dotted, dashdot
plt.plot(batsman['index'], batsman['V Kohli'],color='#03fcf0',linestyle='dashdot')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fc03db',linestyle='dashdot')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
```

```python
#size of line
plt.plot(batsman['index'], batsman['V Kohli'],color='#03fcf0',linewidth=6) #width of line increased
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fc03db')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
```

```python
#To point the marker 
plt.plot(batsman['index'], batsman['V Kohli'],color='#03fcf0', marker='D', markersize=10) # D for diamond | markersize to increase the size of marker
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fc03db',marker='.')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
```

```python
#legend -> location  # legend is  to display the label and location is where to display

plt.plot(batsman['index'], batsman['V Kohli'], label='Kohli',color='black') # label will not display so we use lagend dunction
plt.plot(batsman['index'], batsman['RG Sharma'], label='Rohit')
plt.legend() #to display labels
plt.legend(loc='right')

# location => where to show label following are its perameter
# best => default -> it place the label into vacant area
#upper right
#upper left
#lower right 
#lower left 
# right
# etc
```

```python
#limiting axes  | it is used when one value come very maximum so then  the rest of the graph will look like flat as below
price = [50000,52000, 58000,60000,55000,50000,49000,247000] #last ones's price of iphone is very high
year = [2017,2018,2019,2020,2021,2022,2023,2024] #years
plt.plot(year, price)
#plt.ylim()
plt.ylim(0,90000) # y axes
plt.xlim(2010,2022) # x axes
```

```python
#grid
plt.plot(batsman['index'], batsman['V Kohli'],marker='.')
plt.plot(batsman['index'], batsman['RG Sharma'],marker='.')
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid()
```
